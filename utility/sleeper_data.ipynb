Creates multiple collections
- `draft`: has all data from drafts
- `roster_[season]`: splits draft data by season
    - Running through `transactions` for each year (against their respective `roster_[season]`) should result in matching rosters against `GET roster` endpoint

```python
## Draft data
import requests
import json
from pymongo import MongoClient
from google.colab import userdata
import random
import math

draft_data = {}
season_years = []

def reload_draft_collection_roster_startup():
  db_user = userdata.get('mdb_user')
  db_password = userdata.get('mdb_password')
  db_host = userdata.get('mdb_host')

  client =  MongoClient(f"mongodb+srv://{db_user}:{db_password}@{db_host}/?retryWrites=true&w=majority")
  db = client['nfldata']
  draft_collection = db['draft']

  # reset collections
  draft_collection.delete_many({})
  db['roster_2022'].delete_many({})
  # db['roster_2023'].delete_many({})

  draft_ids = [869408534662696960, 869802031370686464, 1005342622375886848, 966851427416977409]
  draft_ids = [str(draft_id) for draft_id in draft_ids]

  # print(ids[0:1])
  for draft_id in draft_ids:
    get_draft_metadata = requests.get(f"https://api.sleeper.app/v1/draft/{draft_id}")
    draft_metadata = get_draft_metadata.json()
    # print(json.dumps(draft_metadata, indent=2))
    league_id = draft_metadata["league_id"]
    season = int(draft_metadata["season"])
    season_years.append(season)

    # Set the name of the roster_collection to send draft data to
    roster_season = f"roster_{season}"
    roster_collection = db[roster_season]

    # Specify the API endpoint URL
    url = f'https://api.sleeper.app/v1/draft/{draft_id}/picks'

    # Make a GET request
    response = requests.get(url)
    data = response.json()

    if len(data) > 100:
      draft_type = "league"
    else:
      draft_type = "rookie"

    print(f"Successfully pulled draft results: {league_id} | {draft_id} | {season} - {draft_type}")

    # reference
    draft_data[draft_id] = {
        "league_id": league_id,
        "draft_type": draft_type,
        "season": season
    }

    individual_draft_data = []

    for dat in data:
      pick_no = dat["pick_no"]
      cost = 0
      years_left=1
      # needs_contract_status = False

      dat["league_id"]=league_id
      dat["season"]=season
      dat["draft_type"]=draft_type
      # map player and owner data
      player_key = dat["player_id"]
      owner_key = dat["picked_by"]
      dat["team_name"] = users[owner_key]
      dat["player_name"] = players[player_key]["full_name"]
      dat["contract"] = {}
      if draft_type == "rookie":
        years_left=3
        if pick_no <= 5:
          cost = 15
        elif pick_no <= 10:
          cost = 10
        elif pick_no <= 20:
          cost = 5
        elif pick_no <= 30:
          cost = 3
        else:
          cost = 1
        dat["needs_contract_status"] = False
        dat["contract"]["drafted_in"] = season
        dat["contract"]["y0_cost"] = cost
        dat["contract"]["y1_cost"] = int(math.ceil(cost*1.4))
        dat["contract"]["y2_cost"] = int(math.ceil(int(math.ceil(cost*1.4))*1.4))
        dat["contract"]["contract_years_left"] = 3
        dat["contract"]["free_agent_before_season"] = season + years_left

      else:
        # print(dat)
        # print(random.choice(foo))
        dat["needs_contract_status"] = True
        cost = int(dat["metadata"]["amount"])
        dat["contract"]["y0_cost"] = cost
        dat["contract"]["contract_years_left"] = 1
        dat["contract"]["free_agent_before_season"] = season + 1

      # since league draft is set up with rookie draft completed, must remove rookies from league draft pool
      if draft_type == "league" and dat['metadata']['years_exp'] != "0":
        individual_draft_data.append(dat)
      if draft_type == "rookie":
        individual_draft_data.append(dat)
    # only establish roster_2022 collection; subsequent seasons need to be built
    if season==2022:
      roster_collection.insert_many(individual_draft_data)
    # insert draft results from all seasons into single draft collection
    draft_collection.insert_many(individual_draft_data)

reload_draft_collection_roster_startup()
season_years = list(set(season_years))
print(season_years)
```

## Run transaction on rosters (2022 only)
Kickoff season only
- subsequent seasons requires additional processing

```python
import bson
# reload_draft_collection_roster_startup()
# clean_transactions()
client =  MongoClient(f"mongodb+srv://{db_user}:{db_password}@{db_host}/?retryWrites=true&w=majority")
db = client['nfldata']
transactions_collection = db['transactions']

def run_transactions_on_2022_roster():
  # reset rosters
  reload_draft_collection_roster_startup()
  set_contracts_2022()

  # reset transactions
  reload_and_clean_transactions_2022()

  # for season_year in season_years:
  for season_year in [2022]:
    roster_collection = db[f"roster_{season_year}"]

    transactions = list(transactions_collection.find({ "season": season_year}).sort("transaction_id",  1))

    for tx in transactions:
      # If transaction is not a trade, remove owner name and contracts from players that are dropped
      if tx['type'] != 'trade':
        # print("--- not a trade ---")
        # Process drops if present
        if tx['drops_names'] is not None:
          # print(f"< drops >")
          for player, owner in tx['drops_names'].items():
            # print(f"TX: {tx['type']} : {owner} - {player}")
            roster_collection.update_one( { "season":season_year, "player_name": player, "team_name": owner },
                                          { "$unset": { "team_name": 1, "contract": 1 }})

          # For adds, nothing is needed
        if tx['adds_names'] is not None:
          # print(f"< adds >")
          for player, owner in tx['adds_names'].items():
            print(f"TX: {tx['type']} : {owner} - {player}")
            result = roster_collection.update_one( { "season":season_year, "player_name": player },
                                                  { "$set": { "team_name": owner}})
            if result.matched_count == 0:
              # Player not found in roster_collection, fetch from players_collection and add
              # print(f">>> {player}")
              player_data = players_collection.find_one({"full_name": player})
              # print(player_data)
              if player_data:
                  del(player_data['_id'])
                  player_data['season'] = season_year
                  player_data['team_name'] = owner
                  player_data['draft_id'] = "not drafted"
                  player_data['player_name'] = player
                  json.dumps(player_data, indent=2)
                  roster_collection.insert_one(player_data)
                  # roster_collection.update_one( { "season":season_year, "player_name": player },
                  #                                  { "$set": { "team_name": owner}})
                  print(f"{season_year} - {tx['leg']} | {player} added to roster_{ season_year } for team {owner}")
              else:
                  print(f"{season_year} - {tx['leg']} | *** {player} not found in players_collection")


      else:
        # print("------ trade ------")
        # If it is a trade, simply swap owner names on players
        # # For drops, nothing is needed. Owner name will be updated via adds
        if tx['drops_names'] is not None:
          for player, owner in tx['drops_names'].items():
            # print(f"{owner} - {players}")
            roster_collection.update_one( { "season":season_year, "player_name": player },
                                        { "$unset": {"team_name": 1} })

        # For adds, update owner for player with adds
        if tx['adds_names'] is not None:
          # print("< adds >")
          for player, owner in tx['adds_names'].items():
            # print(f"{owner} - {player}")
            roster_collection.update_one( { "season":season_year, "player_name": player },
                                          { "$set": {"team_name": owner} })

run_transactions_on_2022_roster()
```

# Inter-season checklist
At end of the season, a number of things need to happen
1. lock rosters by copying current `roster_[SEASON]` to a new collection `roster_[SEASON]_archive`
2. create `ft` collection and allow owners to franchise tag players (via web tool or spreadsheet)
3. create `roster` collection
  - update `contract` details
    - update `y0_cost` with `y1_cost`, `y1_cost` with `y2_cost`, unset `y2_cost`
    - update `free_agent_before_season` and `contract_years_left`
4. update roster with RFA updates
5. process preseason transactions
  - run through preseason transactions
  - Update roster collection setting up for the draft

6. host draft & update `roster` collection with draft data
  - for each drafted player, check to see if player exists in `roster_collection`
      - if exists, set player `contract` to drafted player
          - set `needs_contract_status` to `false`
      - if does not exist, set `needs_contract_status` to `true`
7. Update roster collection with selected contracts
  - provide `setcontracts` page to set new player contracts

## Checklist #2 - Create a franchise tag collection
Sets the stage for FT page

```python
# from pymongo import PyMongoError

# Rerun this as many times as needed to reset
def create_upcoming_season_collection(next_year):
  completed_season_collection_name = f"roster_{next_year-1}"
  print(completed_season_collection_name)
  client =  MongoClient(f"mongodb+srv://{db_user}:{db_password}@{db_host}/?retryWrites=true&w=majority")
  db = client['nfldata']
  source_collection = db[completed_season_collection_name]
  destination_collection = db[franchise_tag_collection_name]

  ### uncomment this to override the collection protection lock
  destination_collection.delete_many({})

  try:
    check_lock = destination_collection.find_one({"contract.franchise_tag_used": {"$exists": True}})
    if check_lock:
        print("Sorry, cannot delete this collection. Franchise tag information has been gathered")
        return

    destination_collection.delete_many({})

    # Retrieve all documents from the source collection
    documents = list(source_collection.find({ "contract": { "$exists": True } } ) )

    # Insert documents into the destination collection
    if documents:
      for doc in documents:
        doc['contract']['contract_years_left'] = doc['contract']['contract_years_left']
        doc['contract']['free_agent_before_season'] = doc['contract']['free_agent_before_season']
        if 'taxi_designation' in doc['contract']:
          if doc['contract']['taxi_designation']:
            doc['contract']['y3_cost'] = doc['contract']['y2_cost']
            doc['contract']['y2_cost'] = doc['contract']['y1_cost']
            doc['contract']['y1_cost'] = doc['contract']['y0_cost']
            doc['contract']['free_agent_before_season'] = doc['contract']['free_agent_before_season'] + 1
            # contract_years_left
            # free_agent_before_season

      destination_collection.insert_many(documents)
      print(f"Copied {len(documents)} documents from {completed_season_collection_name} to {franchise_tag_collection_name}")
    else:
      print(f"No documents found in {completed_season_collection_name}")

    # destination_collection.update_many({}, {"$set": {"season": next_year}})
    destination_collection.update_many({"contract.contract_years_left": 1}, {"$set": {"contract.franchise_tag_allowed": True}})

  except ConnectionError as e:
    print(f"An error occurred: {e}")

  finally:
    client.close()

create_upcoming_season_collection(next_year=upcoming_season)
```

```python
client =  MongoClient(f"mongodb+srv://{db_user}:{db_password}@{db_host}/?retryWrites=true&w=majority")
db = client['nfldata']
franchise_tag_collection = db[franchise_tag_collection_name]
upcoming_season_collection = db[upcoming_season_collection_name]

def create_roster_collection_from_ft_collection():
  ### uncomment this to override the collection protection lock
  upcoming_season_collection.delete_many({})

  # Only reset collection if lock not present
  try:
    check_lock = upcoming_season_collection.find_one( {"collection_lock": True } )
    if check_lock:
      print("Sorry, cannot delete this collection. Collection is currently locked.\nSee create_roster_collection_from_ft_collection().")
      return

    upcoming_season_collection.delete_many({})

  except ConnectionError as e:
    print(f"An error occurred: {e}")

  # Only get players under contract for next season
  players_under_contract = list(franchise_tag_collection.find({"contract.y1_cost": {"$exists":True}}))

  players_under_contract_list = []
  for player in players_under_contract:
    del(player['_id'])
    player['season'] = upcoming_season
    player['needs_contract_status']=False
    # print(json.dumps(player, indent=2))
    player['contract']['y0_cost'] = player['contract']['y1_cost']
    del(player['contract']['y1_cost'])
    player['contract']['contract_years_left'] = 1
    player['contract']['free_agent_before_season'] = upcoming_season + 1
    if player['contract'].get('y2_cost') is not None:
      player['contract']['y1_cost'] = player['contract']['y2_cost']
      del(player['contract']['y2_cost'])
      player['contract']['contract_years_left'] = 2
      player['contract']['free_agent_before_season'] = upcoming_season + 2
    if player['contract'].get('y3_cost') is not None:
      player['contract']['y2_cost'] = player['contract']['y3_cost']
      del(player['contract']['y3_cost'])
      player['contract']['contract_years_left'] = 3
      player['contract']['free_agent_before_season'] = upcoming_season + 3
    players_under_contract_list.append(player)

  upcoming_season_collection.insert_many(players_under_contract_list)

create_roster_collection_from_ft_collection()
```
